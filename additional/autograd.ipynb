```python
import torch
from torchviz import make_dot
```

```python
def print_var_info(var):
  print("\n----")
  print(f"data: {var.data}")
  print(f"grad: {var.grad}")
  print(f"grad_fn: {var.grad_fn}")
  print(f"is_leaf: {var.is_leaf}")
  print(f"requires_grad: {var.requires_grad}")
  print("----")
```

```python
a = torch.tensor(2.0)
b = torch.tensor(3.0)
```

```python
print_var_info(a)
```

```python
print_var_info(b)
```

```python
c = a * b

print_var_info(c)
```

```python
make_dot(c)
```

Now, let's make `requires_grad = True`

```python
a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(3.0)
```

```python
print_var_info(a)
```

```python
c = a * b
print_var_info(c)
```

```python
make_dot(c)
```

```python
c.backward()
```

```python
print_var_info(c)
```

```python
print_var_info(a)
```
