2. Since deep neural networks has a very huge number of parameters and in a consequence - a huge model complexity - in theory this must lead to the massive overfitting, but in practice - it is the opposite

First of all - regularization is a tecnique that constaints model from being to complex. Early stopping is a great tecnique and it can reduce the training time and keep the model not very complex as it might be. In short: instead of training for huge number of epochs - we can just stop learning when we does not see big improvements in a test error, and this prevents us from having much more complex models with the same generalization error. However, there are cases when traning model for a big number of iterations can lead to the decrease of test error - "double descent".
